### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Ejercicio 1.1 — Promedio, máximo y mínimo sin funciones built-in
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]

# Promedio
total = 0
for valor in esperanzas_de_vida:
    total += valor
promedio = total / len(esperanzas_de_vida)
print(f'Promedio: {promedio:.2f} años')

# Máximo
maximo = esperanzas_de_vida[0]
for valor in esperanzas_de_vida:
    if valor > maximo:
        maximo = valor
print(f'Máximo: {maximo} años')

# Mínimo
minimo = esperanzas_de_vida[0]
for valor in esperanzas_de_vida:
    if valor < minimo:
        minimo = valor
print(f'Mínimo: {minimo} años')
````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Ejercicio 1.3 — Función clasificar_esperanza_vida
def clasificar_esperanza_vida(anios):
    """Clasifica la esperanza de vida en Alta, Media o Baja."""
    if anios >= 75:
        return 'Alta'
    elif anios >= 60:
        return 'Media'
    else:
        return 'Baja'

# Prueba con 5 valores distintos
valores_prueba = [83.2, 74.5, 60.1, 55.0, 49.3]
for v in valores_prueba:
    print(f'{v} años → {clasificar_esperanza_vida(v)}')
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Ejercicio 2.2 — Limpieza de datos

# 1. Rellenar nulos de columnas numéricas con la media de cada columna
columnas_numericas = df.select_dtypes(include='number').columns
for col in columnas_numericas:
    media = df[col].mean()
    df[col] = df[col].fillna(media)

# 2. Verificar que ya no hay nulos
print('Valores nulos después de la limpieza:')
print(df.isnull().sum())
print(f'\nTotal de nulos restantes: {df.isnull().sum().sum()}')
````

**Responde aquí**:

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

> **Respuesta:** Usamos la media porque eliminar filas con nulos reduciría nuestro dataset y podría sesgar el análisis, especialmente si los nulos no son aleatorios. La media conserva la distribución estadística general de la columna y nos permite mantener los 159 países en el análisis. Es una técnica estándar de imputación cuando la cantidad de nulos es pequeña (menos del 10% en este dataset) y los datos no tienen una distribución muy sesgada.

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Ejercicio 2.5 — Agrupaciones con groupby()

# 1. Esperanza de vida promedio por región
print('1. Esperanza de vida promedio por región:')
ev_region = df.groupby('region')['esperanza_vida'].mean()
print(ev_region.round(2))

# 2. Gasto en salud promedio por nivel de ingresos
print('\n2. Gasto en salud promedio por nivel de ingresos:')
gasto_ingresos = df.groupby('nivel_ingresos')['gasto_salud_usd'].mean().sort_values(ascending=False)
print(gasto_ingresos.round(1))

# 3. Número de países por región (continente)
print('\n3. Número de países por continente:')
paises_region = df.groupby('region')['pais'].count()
print(paises_region)

# 4. Mejor y peor región
print(f'\n4. Región con mayor esperanza de vida: {ev_region.idxmax()} ({ev_region.max():.2f} años)')
print(f'   Región con menor esperanza de vida: {ev_region.idxmin()} ({ev_region.min():.2f} años)')
````

### 2.6 — Nueva columna calculada

🔧 **Ejercicio 2.6**
1. Crea una nueva columna llamada `categoria_vida` aplicando la función `clasificar_esperanza_vida` que creaste en la Parte 1 a la columna `esperanza_vida`
2. ¿Cuántos países caen en cada categoría? (usa `value_counts()`)
3. ¿Qué porcentaje del total representa cada categoría?

````python
# Ejercicio 2.6 — Nueva columna calculada

# 1. Crear columna categoria_vida aplicando la función de la Parte 1
df['categoria_vida'] = df['esperanza_vida'].apply(clasificar_esperanza_vida)
print('Primeras filas con la nueva columna:')
display(df[['pais', 'esperanza_vida', 'categoria_vida']].head(10))

# 2. Conteo por categoría
print('\n2. Países por categoría:')
conteo = df['categoria_vida'].value_counts()
print(conteo)

# 3. Porcentaje del total
print('\n3. Porcentaje del total:')
porcentaje = (df['categoria_vida'].value_counts(normalize=True) *100).round(2)
for cat, pct in porcentaje.items(): # categoria cat y pct porcentaje ambos normalizados en la variable de arriba
    print(f'   {cat}: {pct}%')
````

**Escribe tu pregunta:**

**Pregunta formulada:** ¿Existe una relación entre el nivel de ingresos de un país y sus indicadores clave de salud (esperanza de vida, mortalidad infantil, médicos por 1,000 habitantes)?

**Hipótesis:** Los países con mayor nivel de ingresos tienen mejor esperanza de vida, menor mortalidad infantil y más médicos disponibles.


````python
# Ejercicio 2.7 (Opcional) — Relación entre nivel de ingresos e indicadores de salud

# Ordenar los niveles de ingresos de menor a mayor
orden = ['Low', 'Medium', 'High', 'Very High']

# Calcular estadísticas por nivel de ingresos
resumen = df.groupby('nivel_ingresos').agg(
    esperanza_vida_promedio=('esperanza_vida', 'mean'),
    mortalidad_infantil_promedio=('mortalidad_infantil', 'mean'),
    medicos_por_1000_promedio=('medicos_por_1000', 'mean'),
    num_paises=('pais', 'count')
).reindex(orden).round(2)

print('Indicadores de salud promedio por nivel de ingresos:')
display(resumen)

# Análisis de correlación entre gasto en salud y esperanza de vida
correlacion = df['gasto_salud_usd'].corr(df['esperanza_vida'])
print(f'\nCorrelación entre gasto en salud y esperanza de vida: {correlacion:.3f}')

# Hallazgo clave
print('\n--- HALLAZGOS ---')
print(f'Los países "Very High" tienen {resumen.loc["Very High", "esperanza_vida_promedio"]:.1f} años de EV promedio')
print(f'Los países "Low" tienen {resumen.loc["Low", "esperanza_vida_promedio"]:.1f} años de EV promedio')
diferencia = resumen.loc['Very High', 'esperanza_vida_promedio'] - resumen.loc['Low', 'esperanza_vida_promedio']
print(f'Diferencia: {diferencia:.1f} años — el nivel de ingresos tiene un impacto enorme en la salud.')
print(f'La mortalidad infantil en países Low es {resumen.loc["Low", "mortalidad_infantil_promedio"]:.1f}x mayor que en Very High.')
````

### 3.3 — Buenas prácticas

**Responde:**

**¿Qué archivos debería incluir un `.gitignore` en un proyecto de Data Science? Menciona al menos 3 y explica por qué.**

> **Respuesta:**
> 1. `__pycache__/` y `*.pyc` — archivos de caché de Python generados automáticamente; no son parte del código fuente y difieren entre máquinas.
> 2. `.ipynb_checkpoints/` — carpeta que Jupyter crea con copias de seguridad automáticas; no aportan valor al repositorio y aumentan el ruido en los commits.
> 3. `*.env` o `.env` — archivos que contienen variables de entorno como claves de API o contraseñas; subirlos sería una grave vulnerabilidad de seguridad.
> 4. `data/raw/` (datasets grandes) — archivos de datos masivos no deben estar en Git; se gestionan mejor con DVC o almacenamiento en la nube.

**¿Cuál es la diferencia entre `git add .` y `git add nombre_archivo`?**

> **Respuesta:** `git add .` añade TODOS los archivos modificados y nuevos del directorio actual al staging area en un solo comando; es rápido pero puede incluir archivos no deseados. `git add nombre_archivo` añade únicamente ese archivo específico, dando control granular sobre qué cambios se incluyen en el commit; es la práctica recomendada en proyectos donde se quiere mantener commits atómicos y descriptivos.
